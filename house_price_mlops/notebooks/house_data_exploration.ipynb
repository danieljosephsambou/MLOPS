```python
# ðŸ“Œ 1. Import des bibliothÃ¨ques nÃ©cessaires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration des affichages
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

```python
# ðŸ“Œ 2. Chargement du fichier CSV
df = pd.read_csv("../data/house_data.csv")
df.shape
```

```python
# ðŸ“Œ 3. AperÃ§u des premiÃ¨res lignes
df.head()
```

```python
# ðŸ“Œ 4. Informations gÃ©nÃ©rales sur les colonnes
df.info()
```

```python
# ðŸ“Œ 5. Statistiques descriptives des variables numÃ©riques
df.describe()
```

```python
# ðŸ“Œ 6. VÃ©rification des valeurs manquantes
missing_values = df.isnull().sum()
missing_values[missing_values > 0].sort_values(ascending=False)
```

```python
# ðŸ“Œ 7. Pourcentage de valeurs manquantes par colonne
(df.isnull().sum() / len(df) * 100).sort_values(ascending=False)
```

```python
# ðŸ“Œ 8. Visualisation des valeurs manquantes
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="Reds")
plt.title("Carte des valeurs manquantes")
plt.show()
```

```python
# ðŸ“Œ 9. Visualisation de la distribution du prix
plt.figure(figsize=(8, 4))
sns.histplot(df['price'], kde=True)
plt.title("Distribution des prix des maisons")
plt.xlabel("Prix")
plt.ylabel("Nombre")
plt.show()
```

```python
# ðŸ“Œ 10. CorrÃ©lation entre variables numÃ©riques
plt.figure(figsize=(10, 8))
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Matrice de corrÃ©lation")
plt.show()
```

```python
# ðŸ“Œ 11. BoÃ®te Ã  moustaches (outliers) sur quelques variables
variables = ['sqft_living', 'bedrooms', 'yr_built']
for var in variables:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[var])
    plt.title(f"BoÃ®te Ã  moustaches : {var}")
    plt.show()
```
